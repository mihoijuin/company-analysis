# 何を分析するか？？

①まずは顧客データの内容をチェックしよう


・どの国出身者が多いのか

・どの市出身者が多いのか

・どの店舗が利用者が多いのか

```python
# 必要なモジュールをインポート
import sqlite3
import pandas as pd
```

```python
# sakila.dbとのconnectを変数に格納
con = sqlite3.connect('sakila.db')
```

```python
# SQL文の実行内容をDataFrameに返す関数を作る
def sql_to_df(query):
    df = pd.read_sql(query,con)
    
    return df
```

```python
# どの国出身者が多いのか調べる
# customerテーブルとadressテーブル、cityテーブル、countryテーブルを結合し、国ごとの人数を表示する
maincountry_query = ''' SELECT country.country, COUNT(customer_id)
                        FROM customer
                        JOIN address
                            ON customer.address_id = address.address_id
                        JOIN city
                            ON address.city_id = city.city_id
                        JOIN country
                            ON city.country_id = country.country_id 
                        GROUP BY country.country
                        ORDER BY COUNT(customer_id) DESC ; '''
sql_to_df(maincountry_query).head()
```

```python
# どの市出身者が多いのか調べる
maincity_query = ''' SELECT city.city, COUNT(customer_id)
                     FROM customer
                        JOIN address
                            ON customer.address_id = address.address_id
                        JOIN city
                            ON address.city_id = city.city_id
                        GROUP BY city.city
                        ORDER BY COUNT(customer_id) DESC ; '''
sql_to_df(maincity_query).head()
```

```python
# 日本人の出身都市ランキングが見たい
JapaneseCity_query =  ''' SELECT city.city, COUNT(customer_id)
                          FROM customer
                          JOIN address
                              ON customer.address_id = address.address_id
                          JOIN city
                              ON address.city_id = city.city_id
                          JOIN country
                              ON city.country_id = country.country_id
                          WHERE country.country = "Japan"
                          GROUP BY city.city
                          ORDER BY COUNT(customer_id) DESC ; '''

sql_to_df(JapaneseCity_query)
```

```python
# 日本の出身都市カラムを調べてみる
JapaneseCity_query2 = ''' SELECT *
                          FROM customer
                          JOIN address
                              ON customer.address_id = address.address_id
                          JOIN city
                              ON address.city_id = city.city_id
                          JOIN country
                              ON city.country_id = country.country_id
                          WHERE country.country = "Japan" ; '''

sql_to_df(JapaneseCity_query2)
```

```python
# どの店舗の利用者が多いのか調べる'
mainstore_query = ''' SELECT store_id, COUNT(customer_id)
                      FROM customer
                      GROUP BY store_id
                      ORDER BY COUNT(customer_id) DESC ; '''

sql_to_df(mainstore_query)
```
