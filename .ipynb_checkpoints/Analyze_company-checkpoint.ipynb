```python
#モジュールをインポート
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

import sqlite3
```

```python
# sqliteとのコネクトを作る
con = sqlite3.connect('companyAnalysis.db')
```

```python
# 読み込んだDBをDataFrameに格納する関数を作成
def sql_to_df(query):
    
    df = pd.read_sql(query, con)
    return df
```

```python
#接続確認に財務テーブルを取り出す
FS_query = ''' SELECT company_id, name, sales, operating_income, ordinary_income, net_income, cash, operating_CF, investing_CF, financing_CF, stock_price
               FROM financialStatement
               JOIN company
               ON financialStatement.company_id = company.id
               WHERE company_id <> '8527'
               ORDER BY company_id, year'''

FS = sql_to_df(FS_query)
FS
```

```python
# 利益率を求めて新しいカラムに入れる
FS['income_ratio'] = FS['operating_income'] / FS['sales']
```
