```python
#モジュールをインポート
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

import sqlite3
```

```python
# sqliteとのコネクトを作る
con = sqlite3.connect('companyAnalysis.db')
```

```python
# 読み込んだDBをDataFrameに格納する関数を作成
def sql_to_df(query):
    
    df = pd.read_sql(query, con)
    return df
```

```python
#接続確認に財務テーブルを取り出す
FS_query = ''' SELECT company_id, name,year, sales, operating_income, ordinary_income, net_income, cash, operating_CF, investing_CF, financing_CF, stock_price
               FROM financialStatement
               JOIN company
               ON financialStatement.company_id = company.id
               WHERE company_id <> '8527'
               ORDER BY company_id, year'''

FS = sql_to_df(FS_query)
FS
```

```python
# 利益率を求めて新しいカラムに入れる
FS['income_ratio'] = FS['operating_income'] / FS['sales']
```

```python
FS.head(10)
```

```python
#カラム名の取得
column_names = FS.columns
```

```python
column_names
```

```python
FS.iloc[7,3]
```

```python
#モバイルクリエイトのNan値に対応
# 2015年と同数値を入れてみる

# sales
new_sales = FS.iloc[7,3]
FS[column_names[3]] = FS[column_names[3]].fillna(new_sales)


# operating_income
new_op = FS.iloc[7,4]
FS[column_names[4]] = FS[column_names[4]].fillna(new_op)

# ordinary_income
new_or = FS.iloc[7,5]
FS[column_names[5]] = FS[column_names[5]].fillna(new_or)

# net_income
new_net = FS.iloc[7,6]
FS[column_names[6]] = FS[column_names[6]].fillna(new_net)

# cash
new_cash = FS.iloc[7,7]
FS[column_names[7]] = FS[column_names[7]].fillna(new_cash)

# operating_CF
new_oc = FS.iloc[7,8]
FS[column_names[8]] = FS[column_names[8]].fillna(new_oc)

# investing_CF
new_ic = FS.iloc[7,9]
FS[column_names[9]] = FS[column_names[9]].fillna(new_ic)

# financing_CF
new_fc = FS.iloc[7,10]
FS[column_names[10]] = FS[column_names[10]].fillna(new_fc)

# income_ratio
new_ir = FS.iloc[7,12]
FS[column_names[12]] = FS[column_names[12]].fillna(new_ir)
```

```python
#確認
FS.head(10)
```

```python
# とりあえず重回帰分析をする
# 必要なモジュールをインポート
from sklearn.linear_model import LinearRegression as LR
```

```python
# 説明変数に各数値を設定
X = FS[column_names[[3,4,5,6,7,8,9,10]]]
Y = FS['stock_price']
```

```python
# 回帰分析をする
model1 = LR()
model1.fit(X,Y)
```

```python
model1.coef_
```

```python
X.plot()
```

```python
# 企業ごとにやらないとごちゃることが発覚
#企業ごとにDataFrameに分ける
natori = FS[0:5]
mobileC = FS[5:10]
tokyoK = FS[10:15]
termo = FS[15:20]
japanC = FS[20:25]
nikkato = FS[25:30]
hitachiK = FS[30:35]
panasonic = FS[35:40]
apurasF = FS[40:45]
```

```python
# 財務数値と株価をグラフで確認
natori_finances = natori[column_names[[3,4,5,6,7,8,9,10]]]
natoriY = natori['stock_price']

# グラフを見てみる
# 各財務数値
natori_finances.plot()
```

```python
# 株価
natoriY.plot()
```

```python
# 売上を抜いた各指標のグラフ
natori_others = natori[column_names[[4,5,6,7,8,9,10]]]
natori_others.plot()
```

```python
#各数値と株価はある程度影響していそうである
#operating_CFとinvesting_CFは株価と負の関係にありそう

#重回帰モデルで分析してみる
#説明変数は財務諸表の各数値
natoriX = natori[column_names[[3,4,5,6,7,8,9,10]]]

#回帰モデルを作成
model_natori = LR()
model_natori.fit(natoriX, natoriY)
```

```python
# 傾きを確認
model_natori.coef_
```

```python
#めちゃ精度悪そうな数値
#確認のため、2012年のなとりの財務数値で株価を予測してみる

#2012年のDataFrameを作成
natori_2012 =pd.DataFrame([33035, 1465, 1468, 829, 1738, 298, -1342, 1100])

# 分析できる形にするために行と列を入れ替え
natori_2012 =natori_2012.T
```

```python
#予測してみる
model_natori.predict(natori_2012)
```

```python
# 実際なとりの2012年株価は825円とだいぶ違い悲しい結果に（笑）
#一応最初に作った全企業変数からのモデルで確認

pred = model1.predict(natori_2012)
pred
```
